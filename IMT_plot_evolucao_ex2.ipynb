````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
from datetime import datetime
corona_Brasil = pd.read_csv('https://raw.githubusercontent.com/Rogerio-mack/Visualizacao-de-Dados-em-Python/main/data/corona_Brasil.csv')
corona_Brasil.data = pd.to_datetime(corona_Brasil.data, format='%Y-%m-%d',errors='coerce')

bovespa = pd.read_csv('https://raw.githubusercontent.com/Rogerio-mack/Visualizacao-de-Dados-em-Python/main/data/mystocksn.csv')
bovespa.data = pd.to_datetime(bovespa.data, format='%Y-%m-%d',errors='coerce') # somente para ajustar formato da datas

display(corona_Brasil.head())
display(bovespa.head())

# definir data como índice
````

# Exercício 1

Faça um gráfico o mais similar possível do gráfico abaixo.

````python
fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(corona_Brasil.data, corona_Brasil.casosNovos, label='Novos Casos', color='blue')
ax2 = ax.twinx()
ax2.plot(bovespa.data, bovespa.DOLAR, label='DOLAR', color='r')
ax.set_xlabel('date')
ax.set_ylabel('Novos Casos')
ax2.set_ylabel('USD')
ax.legend(loc='upper left')
ax2.legend(loc='upper right')
plt.show()

````

# Exercício 2

Faça um gráfico o mais similar possível do gráfico abaixo.

Dicas:

1. Para a correlação faça o join dos dados por data.
2. Empregue:
```
bovespa.DOLAR.rolling(30).mean()
corona_Brasil.casosNovos.rolling(30).mean()
```
Para gerar as séries médias.


Faça antes, para o ajuste do `rolling()`

````python
corona_Brasil = corona_Brasil.set_index('data')
bovespa = bovespa.set_index('data')
````

````python
fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(corona_Brasil.index, corona_Brasil.casosNovos, label='Novos Casos', color='blue',alpha=0.3)
ax2 = ax.twinx()
ax2.plot(bovespa.index, bovespa.DOLAR, label='DOLAR', color='r',alpha=0.3)
ax.set_xlabel('date')
ax.set_ylabel('Novos Casos')
ax2.set_ylabel('USD')
ax.legend(loc='upper left')
ax2.legend(loc='upper right')
ax.plot(corona_Brasil.index, corona_Brasil.casosNovos.rolling(30).mean(), label='Novos Casos Média', color='blue')
ax2.plot(bovespa.index, bovespa.DOLAR.rolling(30).mean(), label='DOLAR Média', color='r')
# fazer corr
corr = corona_Brasil.casosNovos.corr(bovespa.DOLAR)
print(corr)
# annotate
ax.text(0.4, 0.8, f'Correlation = {corr:.2f}', transform=ax.transAxes, fontsize=20, color='darkred')
# ax.title('Novos Casos de Covid x USD (Brasil)')
plt.show()
````
